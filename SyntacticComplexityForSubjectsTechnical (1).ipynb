```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import matplotlib.pyplot as plt
import itertools
from scipy.stats import ttest_ind

file_path = "/content/drive/My Drive/Colab Notebooks/14-Indices (2).txt"

W = {"Lay Summary" : {}, "Technical Summary": {} }
S = {"Lay Summary" : {}, "Technical Summary": {} }
VP = {"Lay Summary" : {}, "Technical Summary": {} }
C = {"Lay Summary" : {}, "Technical Summary": {} }
T = {"Lay Summary" : {}, "Technical Summary": {} }
DC = {"Lay Summary" : {}, "Technical Summary": {} }
CT = {"Lay Summary" : {}, "Technical Summary": {} }
CP = {"Lay Summary" : {}, "Technical Summary": {} }
CN = {"Lay Summary" : {}, "Technical Summary": {} }
MLS = {"Lay Summary" : {}, "Technical Summary": {} }
MLT = {"Lay Summary" : {}, "Technical Summary": {} }
MLC = {"Lay Summary" : {}, "Technical Summary": {} }
C_S = {"Lay Summary" : {}, "Technical Summary": {} }
VP_T = {"Lay Summary" : {}, "Technical Summary": {} }
C_T = {"Lay Summary" : {}, "Technical Summary": {} }
DC_C = {"Lay Summary" : {}, "Technical Summary": {} }
DC_T = {"Lay Summary" : {}, "Technical Summary": {} }
T_S = {"Lay Summary" : {}, "Technical Summary": {} }
CT_T = {"Lay Summary" : {}, "Technical Summary": {} }
CP_T = {"Lay Summary" : {}, "Technical Summary": {} }
CP_C = {"Lay Summary" : {}, "Technical Summary": {} }
CN_T = {"Lay Summary" : {}, "Technical Summary": {} }
CN_C = {"Lay Summary" : {}, "Technical Summary": {} }

subjectLabels = {}

MasterDictionary2 = {"W" : W, "S" : S, "VP": VP, "C": C, "T": T, "DC" : DC, "CT" : CT, "CP" :CP, "CN" :CN, "MLS" : MLS, "MLT" : MLT, "MLC" : MLC, "C_S" : C_S, "VP_T" : VP_T, "C_T" : C_T, "DC_C" : DC_C, "DC_T" : DC_T, "T_S" : T_S, "CT_T" : CT_T, "CP_T" : CP_T, "CP_C" : CP_C, "CN_T" : CN_T, "CN_C" : CN_C}
Subjects = {}

lineNum = 0
laySummaryCount = 0
techSummaryCount = 0

with open(file_path, "r") as file:

    for line in file:

        #Document all the subjects of the summaries
        lineNum += 1
        if lineNum % 3 == 1:
          subject = line.strip().lower()
          if subject in Subjects.keys():
            Subjects[subject] += 1
          else:
            Subjects[subject] = 1
        else:
          # Remove trailing newline and whitespace
          preValues = [val.strip() for val in line.strip().split(",")]
          values = []
          for value in preValues:
            values.append(float(value))

          if lineNum % 3 == 2:

            laySummaryCount += 1
            subjectLabels[laySummaryCount] = subject.strip().lower()

          else:

            techSummaryCount += 1

          for value in values:


            if lineNum % 3 == 2:


              W["Lay Summary"][laySummaryCount] = values[0]
              S["Lay Summary"][laySummaryCount] = values[1]
              VP["Lay Summary"][laySummaryCount] = values[2]
              C["Lay Summary"][laySummaryCount] = values[3]
              T["Lay Summary"][laySummaryCount] = values[4]
              DC["Lay Summary"][laySummaryCount] = values[5]
              CT["Lay Summary"][laySummaryCount] = values[6]
              CP["Lay Summary"][laySummaryCount] = values[7]
              CN["Lay Summary"][laySummaryCount] = values[8]
              MLS["Lay Summary"][laySummaryCount] = values[9]
              MLT["Lay Summary"][laySummaryCount] = values[10]
              MLC["Lay Summary"][laySummaryCount] = values[11]
              C_S["Lay Summary"][laySummaryCount] = values[12]
              VP_T["Lay Summary"][laySummaryCount] = values[13]
              C_T["Lay Summary"][laySummaryCount] = values[14]
              DC_C["Lay Summary"][laySummaryCount] = values[15]
              DC_T["Lay Summary"][laySummaryCount] = values[16]
              T_S["Lay Summary"][laySummaryCount] = values[17]
              CT_T["Lay Summary"][laySummaryCount] = values[18]
              CP_T["Lay Summary"][laySummaryCount] = values[19]
              CP_C["Lay Summary"][laySummaryCount] = values[20]
              CN_T["Lay Summary"][laySummaryCount] = values[21]
              CN_C["Lay Summary"][laySummaryCount] = values[22]
            else:
              W["Technical Summary"][techSummaryCount] = values[0]
              S["Technical Summary"][techSummaryCount] = values[1]
              VP["Technical Summary"][techSummaryCount] = values[2]
              C["Technical Summary"][techSummaryCount] = values[3]
              T["Technical Summary"][techSummaryCount] = values[4]
              DC["Technical Summary"][techSummaryCount] = values[5]
              CT["Technical Summary"][techSummaryCount] = values[6]
              CP["Technical Summary"][techSummaryCount] = values[7]
              CN["Technical Summary"][techSummaryCount] = values[8]
              MLS["Technical Summary"][techSummaryCount] = values[9]
              MLT["Technical Summary"][techSummaryCount] = values[10]
              MLC["Technical Summary"][techSummaryCount] = values[11]
              C_S["Technical Summary"][techSummaryCount] = values[12]
              VP_T["Technical Summary"][techSummaryCount] = values[13]
              C_T["Technical Summary"][techSummaryCount] = values[14]
              DC_C["Technical Summary"][techSummaryCount] = values[15]
              DC_T["Technical Summary"][techSummaryCount] = values[16]
              T_S["Technical Summary"][techSummaryCount] = values[17]
              CT_T["Technical Summary"][techSummaryCount] = values[18]
              CP_T["Technical Summary"][techSummaryCount] = values[19]
              CP_C["Technical Summary"][techSummaryCount] = values[20]
              CN_T["Technical Summary"][techSummaryCount] = values[21]
              CN_C["Technical Summary"][techSummaryCount] = values[22]


physicalSciences = ['environmental sciences', 'earth, atmospheric, and planetary sciences', 'applied physical sciences', 'computer sciences','applied mathematics','engineering','chemistry','statistics','sustainability science','physics','astronomy']
biologicalSciences = ['biophysics and computational biology','immunology and inflammation','neuroscience', 'genetics', 'ecology', 'medical sciences','evolution','plant biology','agricultural sciences','applied biological sciences','biochemistry','developmental biology','systems biology','microbiology','pharmacology','cell biology','physiology','population biology']
socialSciences = ['economic sciences','psychological and cognitive sciences','demography','political sciences','social sciences','anthropology','']

umbrellas = {'physicalSciences':physicalSciences,'biologicalSciences':biologicalSciences,'socialSciences': socialSciences}

physCount = 0
bioCount = 0
socCount = 0

index = 0

for index in subjectLabels.keys():

  subject = subjectLabels[index]

  if subject in physicalSciences:

    physCount += 1

  elif subject in biologicalSciences:

    bioCount += 1

  elif subject in socialSciences:

    socCount += 1

bioQ1Q2Mean = {"W":{'q1':0,'q3':0, 'median':0}, "S":{'q1':0,'q3':0, 'median':0}, "VP":{'q1':0,'q3':0, 'median':0}, "C":{'q1':0,'q3':0, 'median':0}, "T":{'q1':0,'q3':0, 'median':0}, "DC":{'q1':0,'q3':0, 'median':0}, "CT":{'q1':0,'q3':0, 'median':0}, "CP":{'q1':0,'q3':0, 'median':0}, "CN":{'q1':0,'q3':0, 'median':0},
    "MLS":{'q1':0,'q3':0, 'median':0}, "MLT":{'q1':0,'q3':0, 'median':0}, "MLC":{'q1':0,'q3':0, 'median':0}, "C_S":{'q1':0,'q3':0, 'median':0}, "VP_T":{'q1':0,'q3':0, 'median':0}, "C_T":{'q1':0,'q3':0, 'median':0}, "DC_C":{'q1':0,'q3':0, 'median':0}, "DC_T":{'q1':0,'q3':0, 'median':0},
    "T_S":{'q1':0,'q3':0, 'median':0}, "CT_T":{'q1':0,'q3':0, 'median':0}, "CP_T":{'q1':0,'q3':0, 'median':0}, "CP_C":{'q1':0,'q3':0, 'median':0}, "CN_T":{'q1':0,'q3':0, 'median':0}, "CN_C":{'q1':0,'q3':0, 'median':0}}
physQ1Q2Mean = {"W":{'q1':0,'q3':0, 'median':0}, "S":{'q1':0,'q3':0, 'median':0}, "VP":{'q1':0,'q3':0, 'median':0}, "C":{'q1':0,'q3':0, 'median':0}, "T":{'q1':0,'q3':0, 'median':0}, "DC":{'q1':0,'q3':0, 'median':0}, "CT":{'q1':0,'q3':0, 'median':0}, "CP":{'q1':0,'q3':0, 'median':0}, "CN":{'q1':0,'q3':0, 'median':0},
    "MLS":{'q1':0,'q3':0, 'median':0}, "MLT":{'q1':0,'q3':0, 'median':0}, "MLC":{'q1':0,'q3':0, 'median':0}, "C_S":{'q1':0,'q3':0, 'median':0}, "VP_T":{'q1':0,'q3':0, 'median':0}, "C_T":{'q1':0,'q3':0, 'median':0}, "DC_C":{'q1':0,'q3':0, 'median':0}, "DC_T":{'q1':0,'q3':0, 'median':0},
    "T_S":{'q1':0,'q3':0, 'median':0}, "CT_T":{'q1':0,'q3':0, 'median':0}, "CP_T":{'q1':0,'q3':0, 'median':0}, "CP_C":{'q1':0,'q3':0, 'median':0}, "CN_T":{'q1':0,'q3':0, 'median':0}, "CN_C":{'q1':0,'q3':0, 'median':0}}
sociQ1Q2Mean = {"W":{'q1':0,'q3':0, 'median':0}, "S":{'q1':0,'q3':0, 'median':0}, "VP":{'q1':0,'q3':0, 'median':0}, "C":{'q1':0,'q3':0, 'median':0}, "T":{'q1':0,'q3':0, 'median':0}, "DC":{'q1':0,'q3':0, 'median':0}, "CT":{'q1':0,'q3':0, 'median':0}, "CP":{'q1':0,'q3':0, 'median':0}, "CN":{'q1':0,'q3':0, 'median':0},
    "MLS":{'q1':0,'q3':0, 'median':0}, "MLT":{'q1':0,'q3':0, 'median':0}, "MLC":{'q1':0,'q3':0, 'median':0}, "C_S":{'q1':0,'q3':0, 'median':0}, "VP_T":{'q1':0,'q3':0, 'median':0}, "C_T":{'q1':0,'q3':0, 'median':0}, "DC_C":{'q1':0,'q3':0, 'median':0}, "DC_T":{'q1':0,'q3':0, 'median':0},
    "T_S":{'q1':0,'q3':0, 'median':0}, "CT_T":{'q1':0,'q3':0, 'median':0}, "CP_T":{'q1':0,'q3':0, 'median':0}, "CP_C":{'q1':0,'q3':0, 'median':0}, "CN_T":{'q1':0,'q3':0, 'median':0}, "CN_C":{'q1':0,'q3':0, 'median':0}}


for key in MasterDictionary2.keys():
  bio = []
  phys = []
  soci = []

  for i in range(1,301):

    if subjectLabels[i] in physicalSciences:
      phys.append(MasterDictionary2[key]["Technical Summary"][i])
    elif subjectLabels[i] in biologicalSciences:
      bio.append(MasterDictionary2[key]["Technical Summary"][i])
    elif subjectLabels[i] in socialSciences:
      soci.append(MasterDictionary2[key]["Technical Summary"][i])

  bioQ1Q2Mean[key]['q1'] = np.percentile(np.fromiter(bio, dtype=float),25)
  bioQ1Q2Mean[key]['q3'] = np.percentile(np.fromiter(bio, dtype=float),75)
  bioQ1Q2Mean[key]['mean'] = np.median(np.fromiter(bio,dtype=float))

  physQ1Q2Mean[key]['q1'] = np.percentile(np.fromiter(phys, dtype=float),25)
  physQ1Q2Mean[key]['q3'] = np.percentile(np.fromiter(phys, dtype=float),75)
  physQ1Q2Mean[key]['median'] = np.median(np.fromiter(phys,dtype=float))

  sociQ1Q2Mean[key]['q1'] = np.percentile(np.fromiter(soci, dtype=float),25)
  sociQ1Q2Mean[key]['q3'] = np.percentile(np.fromiter(soci, dtype=float),75)
  sociQ1Q2Mean[key]['median'] = np.median(np.fromiter(soci,dtype=float))

Measures = [W,S,VP,C,T,DC,CT,CP,CN,MLS,MLT,MLC,C_S,VP_T,C_T,DC_C,DC_T,T_S,CT_T,CP_T,CP_C,CN_T,CN_C]

measure_names = [
    "W", "S", "VP", "C", "T", "DC", "CT", "CP", "CN",
    "MLS", "MLT", "MLC", "C_S", "VP_T", "C_T", "DC_C", "DC_T",
    "T_S", "CT_T", "CP_T", "CP_C", "CN_T", "CN_C"
]


results3 = {}
for pair in itertools.combinations(umbrellas.keys(),2):
  results3[pair] = {}

for pair in itertools.combinations(umbrellas.keys(), 2):

  subjOne = pair[0]
  subjTwo = pair[1]

  for name, measure in zip(measure_names, Measures):

    subOneValues = []
    subTwoValues = []


    for i in range(1,300):

      if subjectLabels[i] in umbrellas[subjOne]:
        subOneValues.append(measure["Technical Summary"][i])

      elif subjectLabels[i] in umbrellas[subjTwo]:

        subTwoValues.append(measure["Technical Summary"][i])

    t_stat, p_val = ttest_ind(subOneValues, subTwoValues)

    results3[pair][name] = {'t': t_stat, 'p': p_val}

    print(f"{name} for {pair}")

    plt.hist(subOneValues, bins = 30, color = 'skyblue', alpha = 0.5, label = "lay")

    plt.hist(subTwoValues, bins = 30, color = 'red', alpha = 0.5, label = "tech")

    plt.show()

    plt.ecdf(subOneValues, color = 'skyblue', alpha = 0.5, label = "lay")

    plt.ecdf(subTwoValues, color = 'red', alpha = 0.5, label = "tech")

    plt.show()

for pair in results3.keys():

  for measure in measure_names:
    if results3[pair][measure]['p']<0.05:
      print(f"significant different in {measure} for {pair}, t-value = {results3[pair][measure]['t']}, p-value = {results3[pair][measure]['p']} ")
    elif results3[pair][measure]['p']<0.1:
      print(f"weak difference in {measure} for {pair}, t-value = {results3[pair][measure]['t']}, p-value = {results3[pair][measure]['p']} ")
    elif results3[pair][measure]['p']>=0.1:
      print(f"no significant difference in {measure} for {pair}, t-value = {results3[pair][measure]['t']}, p-value = {results3[pair][measure]['p']} ")

#print(bioQ1Q2Mean)
#print(physQ1Q2Mean)
#print(sociQ1Q2Mean)
#print(results3)
```
